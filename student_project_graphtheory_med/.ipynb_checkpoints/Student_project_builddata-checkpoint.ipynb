```python
article_stats = con.execute("""
    SELECT
        page_id,
        COUNT(*) AS total_edits,
        COUNT(DISTINCT event_user_text_historical) AS num_editors
    FROM no_bots
    WHERE page_namespace = 0
      AND page_is_redirect = FALSE
    GROUP BY page_id
""").fetchdf()

# Create bins
def editor_bin(n):
    if n <= 2:
        return "1–2 editors"
    elif n <= 5:
        return "3–5 editors"
    elif n <= 10:
        return "6–10 editors"
    elif n <= 50:
        return "11–50 editors"
    else:
        return "50+ editors"

article_stats["editor_bin"] = article_stats["num_editors"].apply(editor_bin)

bin_counts = article_stats["editor_bin"].value_counts().sort_index()

bin_counts
```

```python
# Article-level stats (after removing bots and redirects)

con.execute("""
CREATE OR REPLACE TABLE article_stats AS
SELECT
    page_id,
    page_title,
    COUNT(*) AS total_edits,
    COUNT(DISTINCT event_user_text_historical) AS num_editors
FROM no_bots
WHERE page_namespace = 0
  AND page_is_redirect = FALSE
GROUP BY page_id, page_title
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE filtered_articles AS
SELECT *
FROM article_stats
WHERE num_editors >= 2
  AND total_edits >= 10
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE sample_articles AS
SELECT page_id
FROM filtered_articles
USING SAMPLE 800 ROWS
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE editor_article_sample AS
SELECT
    nb.page_id,
    nb.page_title AS article,
    nb.event_user_text_historical AS editor,
    COUNT(*) AS num_edits
FROM no_bots nb
JOIN sample_articles sa
  ON nb.page_id = sa.page_id
WHERE nb.page_namespace = 0
  AND nb.page_is_redirect = FALSE
GROUP BY nb.page_id, nb.page_title, nb.event_user_text_historical
""")
```

```python
con.execute("""
SELECT COUNT(*) AS bot_rows
FROM editor_article_sample eas
JOIN no_bots nb
  ON eas.page_id = nb.page_id
 AND eas.editor = nb.event_user_text_historical
WHERE nb.event_user_is_bot_by IS NOT NULL
""").fetchdf()
```

```python
con.execute("""
SELECT COUNT(*) AS redirect_rows
FROM editor_article_sample eas
JOIN no_bots nb
  ON eas.page_id = nb.page_id
WHERE nb.page_is_redirect = TRUE
""").fetchdf()
```

```python
con.execute("""
SELECT COUNT(*) AS non_ns0_rows
FROM editor_article_sample eas
JOIN no_bots nb
  ON eas.page_id = nb.page_id
WHERE nb.page_namespace <> 0
""").fetchdf()
```

```python
con.execute("""
SELECT
    MIN(num_editors) AS min_editors,
    MIN(total_edits) AS min_total_edits
FROM (
    SELECT
        page_id,
        COUNT(DISTINCT editor) AS num_editors,
        SUM(num_edits) AS total_edits
    FROM editor_article_sample
    GROUP BY page_id
)
""").fetchdf()
```

```python
con.execute("""
SELECT COUNT(DISTINCT page_id) AS n_articles
FROM editor_article_sample
""").fetchdf()
```

```python
# Article-level stats after removing bots, redirects, namespace != 0
con.execute("""
CREATE OR REPLACE TABLE article_stats AS
SELECT
    page_id,
    page_title,
    COUNT(*) AS total_edits,
    COUNT(DISTINCT event_user_text_historical) AS num_editors
FROM no_bots
WHERE page_namespace = 0
  AND page_is_redirect = FALSE
GROUP BY page_id, page_title
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE filtered_articles AS
SELECT *
FROM article_stats
WHERE num_editors >= 2
  AND total_edits >= 10
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE sample_articles AS
SELECT page_id
FROM filtered_articles
USING SAMPLE 800 ROWS
""")
```

```python
con.execute("""
CREATE OR REPLACE TABLE revision_sample AS
SELECT nb.*
FROM no_bots nb
JOIN sample_articles sa
  ON nb.page_id = sa.page_id
WHERE nb.page_namespace = 0
  AND nb.page_is_redirect = FALSE
""")
```

```python
con.execute("""
SELECT 
    COUNT(*) AS n_rows,
    COUNT(DISTINCT page_id) AS n_articles,
    COUNT(DISTINCT event_user_text_historical) AS n_editors
FROM revision_sample
""").fetchdf()
```
